## Data Cleaning

#### Outlier Detection

**Let's identify and drop outliers, once both Knn and SVM are sensitive to those.\
For out purposes, outliers will be defined as values more than 1.5\*IQR (*inter-quartile range* |Q3-Q1|) above the the third quartile or below the first quartile.**

**We'll start by defining three helper methods:**
- One to calculate outliers for a given column
- Another to give a summary of ouliers present on the dataframe
- And finally, one to drop the rows containing outliers

```python
def get_outliers(df, col):
    '''
    Method that given a dataframe and a column, calculates Interquartile Range
    identifies outiliers in the dataframe and returns them.
    '''
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    # identify outliers
    threshold = 1.5
    outliers = df[(df[col] < Q1 - threshold * IQR) | (df[col] > Q3 + threshold * IQR)]
    return outliers


def get_outliers_summary(df):
    '''
    Given a dataframe, return another dataframe with a basic summary
    of outliers
    '''
    outliers_info = list()
    for col in df.drop('color', axis=1).columns:
        outliers = get_outliers(df, col)
        if not outliers.empty:
            outliers_info.append({
                'column': col,
                'Nº of outliers': outliers.shape[0]
            })
    return pd.DataFrame(outliers_info)
 
    
def drop_outliers(df):
    '''
    Given a dataframe, returns a df with no outlier rows
    '''
    for col in df.drop('color', axis=1).columns:
        outliers = get_outliers(df, col)
        df = df.drop(outliers.index)
    return df
```

### GridSearch CV Comparison with manual Hyperparameter Tuning

We'll define a manual hyperparameter tuning method and compare the results obtained with the one from GridSearch CV

```python
def get_best_knn_params(X_train, y_train,X_test, y_test, param_space):
    '''
    Method that finds optimal hyperparameters for knn given train/testdata and a dict with
    the parameter space for training
    Returns a pandas dataframe with a summary of the best params and scores
    '''
    best_params = list()
    max_score = 0
    neighbors_range = param_space['n_neighbors']
    powers = param_space['p']
    weights = param_space['weights']

    for n in neighbors_range:
        for w in weights:
            for p in powers:
                knn = KNeighborsClassifier(n_neighbors=n, weights=w, p=p)
                knn.fit(X_train, y_train)
                score = knn.score(X_test, y_test)

                if score > max_score:
                    max_score = score
                    best_params.append({
                        'n_neighbors': n,
                        'weight': w,
                        'p': p,
                        'test_score': round(score, 3)
                    })

    bets_params_df = pd.DataFrame(best_params)
    pretty_print_df(bets_params_df)
    return bets_params_df
```

```python
parameters = {'n_neighbors':[i for i in range(2, 50, 2)], 'weights':('uniform', 'distance'), 'p': [1,2]}
```

```python
best_params = get_best_knn_params(X_train, y_train,X_test, y_test, parameters)
```

```python
t0_knn = time.time()
knn = KNeighborsClassifier(n_neighbors=40, weights='distance', p=1)
knn.fit(X_train, y_train)
score = knn.score(X_test, y_test)
tf_knn = time.time()-t0_knn

print('Training time:', round(tf_knn*1000, 2), 'ms')
print('Accuracy Score:', round(score, 3))
```

### Comparison with manual  Tuning

```python
def get_best_svm_params(X_train, y_train,X_test, y_test, param_space):
    '''
    Method that finds optimal hyperparameters for SVM given train/test data and a dict with
    the parameter space for training
    Returns a pandas dataframe with a summary of the best params and scores
    '''
    best_params = list()
    max_score = 0
    Cs = param_space['C']
    gammas = param_space['gamma']
    
    for c in Cs:
        for g in gammas:
            svc = SVC(kernel='rbf', C=c, gamma=g)
            svc.fit(X_train, y_train)
            score = accuracy_score(svc.predict(X_test), y_test)

            if score > max_score:
                max_score = score
                best_params.append({
                    'C': c,
                    'Gamma': g,
                    'test_score': round(score, 3)
                })

    bets_params_df = pd.DataFrame(best_params)
    pretty_print_df(bets_params_df)
    return bets_params_df
```

```python
parameters = {
    'C':[i for i in range(1,10,2)],
    'gamma':[i for i in range(10,20,2)]
}
```

```python
best_params_svm = get_best_svm_params(X_train, y_train,X_test, y_test, parameters)
```

```python
t0_svc = time.time()
svc = SVC(kernel='rbf', C=7, gamma=16)
svc.fit(X_train, y_train)
tf_svc = time.time()-t0_svc
score = round(accuracy_score(svc.predict(X_test), y_test), 3)

print('Training time:', round(tf_svc*1000, 2), 'ms')
print('Accuracy Score:', round(score, 3))
```
